```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Telco customer.csv')
```

```python
df.columns
```

```python
df['PaymentMethod'].value_counts()
```

```python
df.drop('customerID', axis=1, inplace=True)
df.drop('gender', axis=1, inplace=True)
```

```python
df.isnull().sum()
```

```python
df = df.applymap(lambda x: 0 if isinstance(x, str) and x.strip().lower().startswith('no') else x)
df = df.applymap(lambda x: 1 if isinstance(x, str) and x.strip().lower().startswith('yes') else x)
```

```python
df = pd.get_dummies(df, columns=['InternetService'], drop_first=True)
df = pd.get_dummies(df, columns=['PaymentMethod'], drop_first=True)
df['Contract'] = df['Contract'].map({'Month-to-month': 1, 'One year': 12, 'Two year': 24})
```

```python
df
```

```python
# it will convert the non-numeric values to NaN and all other remains the same
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
df.dropna(inplace=True)
```

```python
df.dtypes
```

```python
for i in df.columns:
    if df[i].dtype == 'bool':
        df[i] = df[i].astype(int)
df
```

```python
# outliers
a =['tenure', 'MonthlyCharges', 'TotalCharges']
for i in a:
  Q1 = df[i].quantile(0.25)
  Q3 = df[i].quantile(0.75)
  IQR = Q3 - Q1
  outliers = ((df[i] < (Q1 - 1.5 * IQR)) |(df[i] > (Q3 + 1.5 * IQR)))
  print(outliers.sum())
```

```python
service_cols = [
    'OnlineSecurity',
    'OnlineBackup',
    'DeviceProtection',
    'TechSupport',
    'StreamingTV',
    'StreamingMovies'
]

df['ServiceCount'] = 0
for i in service_cols:
    df['ServiceCount']+= df[i]
df
```

# **Modeling**

## Splitting

```python
x = df.drop('Churn', axis=1)
y = df['Churn']
```

```python
# standard
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x = scaler.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
#smoting
from imblearn.over_sampling import SMOTE
smote = SMOTE()
x_train, y_train = smote.fit_resample(x_train, y_train)
```

## Logistic Regression



```python
# logistic regression
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
from sklearn.metrics import classification_report
y_pred = model.predict(x_test)
print(classification_report(y_test, y_pred))
```

```python
# confusion matrix
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

### Multinominal Logistic Regression

```python
# Multinominal Logistic Regression
model = LogisticRegression(multi_class='multinomial')
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
y_pred = model.predict(x_test)
print(classification_report(y_test, y_pred))
```

```python
# confusion matrix
confusion_matrix(y_test, y_pred)
```

## KNN

```python
# knn
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=15)
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
y_pred = model.predict(x_test)
print(classification_report(y_test, y_pred))
```

```python
# confusion matrix
confusion_matrix(y_test, y_pred)
```

## SVM

```python
# svm
from sklearn.svm import SVC
model = SVC(kernel='linear')
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
y_pred = model.predict(x_test)
print(classification_report(y_test, y_pred))
```

```python
# confusion matrix
confusion_matrix(y_test, y_pred)
```
